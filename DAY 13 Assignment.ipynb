```
import numpy as np

x = np.array(cards['Time'])
y = np.array(cards['Amount'])

# Ploting the graph
print(plt.plot(x,y))

# Labling the graph
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

```
import numpy as np   ## 6. LINE Space plot
x = np.linspace(0, 20, 1000)

fig = plt.figure()
plt.plot(x, np.sin(x), '-', 'r')
plt.plot(x, np.cos(x), '--', 'y')
print(plt.title('Class VS Amount'))
plt.show()
```

```
sns.boxplot(x='Class',y='Amount',data=cards)   ##BOX PLOT
print(plt.title('Class VS Amount'))
plt.show()
```

```
plt.figure(figsize=(12,6))
sns.boxplot(x='Class',y='Amount',data=cards)
plt.ylim(0,1000)
print(plt.title('Class VS Amount'))
plt.show()
```

```
sns.barplot(x='Class',y='Amount',data=cards)  ##BAR PLOT
print(plt.title('Class VS Amount'))
plt.show()
```

```
sns.scatterplot(x='Class',y='Amount',data=cards)  ##SCATTER PLOT
print(plt.title('Class VS Amount'))
plt.show()
```
