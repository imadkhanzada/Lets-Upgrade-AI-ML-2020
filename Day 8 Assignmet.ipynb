```
## using map function
def addition(n): 
    return n + n 
numbers = (9, 8, 7, 6) 
result = map(addition, numbers) 
print(list(result)) 
```

```

#using Lamba function
l= [9,8,7,6]
x= map(lambda x:x+2,l)
print(list(x))
```

```
x = lambda a, b, c, d: a + b + c + d     #using Lamba function
print(x(9, 8, 7,6))
```

```
#using  filter function
l= [1,2,4,9,8,7,6]
x= filter(lambda x:x>=2,l)
print(list(x))
```

```
# A Sample class with init function

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Imad", 25)

print("my name is ",p1.name)
print("my age is",p1.age)
```
