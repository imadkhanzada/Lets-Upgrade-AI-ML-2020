**Question 1:**
Create one array of actual values and another array of predicted values. Compare the two sets
with the confusion matrix.



```
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```
##Actual Values 
actual=[2,0,2,0,2,2,0,2,2,2]
##Predicted Values 
predicted=[0,2,2,2,0,2,0,2,0,2]
```

```
##Confusion matrix
matrix=confusion_matrix(actual,predicted,labels=[2,0])
print('confusion matrix:\n',matrix)

# outcome values order in sklearn
tp , fn, fp, tn =confusion_matrix(actual,predicted,labels=[2,0]).reshape(-1)
print('Outcome values : \n', tp, fn, fp, tn)

# classification report for precision, recall f1-score and accuracy
matrix=classification_report(actual,predicted,labels=[2,0])
print('Classification report : \n',matrix)
```

**Question 2:**
Find out the recall, precision, F1 score and confusion matrix with picture

1.  cat(42,8)
2.   dog(18,32)



```
##confusion mtrix 
matrix=np.array([[42,8],[18,32]])
print('confusion matrix:\n',matrix)
# True positive
tp = matrix[0, 0]
# True negetive
tn = matrix[1, 1]
# False positive
fp = matrix[0, 1]
# False negetive
fn = matrix[1, 0]
##Accuracy Score
accuracy=(tp+tn)/(tp+tn+fp+fn)
print('\nAccuracy=',accuracy)
##Recall
recall=(tp/(tp+fn))
print('\n Recall=',recall)
##Precision
precision=(tp/(tp+fp))
print('\n Precision=',precision)
##f1 score 
f1_score=((2*(precision*recall))/(precision+recall))
print('\n f1 score=',f1_score)
```
